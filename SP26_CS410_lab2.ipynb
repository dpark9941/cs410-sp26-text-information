#PageRank reference

```python
import numpy as np
```

```python
def run_pagerank(graph, d=0.85, iters=20):
    nodes = list(graph.keys())
    n = len(nodes)
    scores = {node: 1/n for node in nodes}

    for _ in range(iters):
        new_scores = {node: (1 - d) / n for node in nodes}
        for node, links in graph.items():
            if links:
                outbound_share = scores[node] / len(links)
                for link in links:
                    if link in new_scores:
                        new_scores[link] += d * outbound_share
        scores = new_scores

    # Sort by score to see the "winners"
    return dict(sorted(scores.items(), key=lambda x: x[1], reverse=True))
```
